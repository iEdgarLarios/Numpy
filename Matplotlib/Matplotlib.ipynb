```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movie_metadata.csv', nrows=20, skiprows=[1])
```

```python
movies.head(3)
```

```python
movies.groupby('genres')['movie_facebook_likes'].sum().plot(kind='line', legend='Reverse')
```

```python
movies.groupby('genres')['movie_facebook_likes'].sum().plot(kind='pie', legend='Reverse')
```

```python
movies.groupby('genres')['movie_facebook_likes'].sum().plot(kind='bar', legend='Reverse')
```

```python
y = movies['duration']
x = movies['genres']

plt.xlabel('genres')
plt.ylabel('duration')
plt.scatter(x,y)
```

```python
y = movies['duration']
x = movies['genres']

plt.xlabel('genres')
plt.ylabel('duration')
plt.bar(x,y)
```

```python
movies.num_critic_for_reviews.groupby(movies.genres).sum().plot(kind='pie', cmap='Paired')
```

```python
movies.num_critic_for_reviews.groupby(movies.genres).sum().plot(kind='bar', cmap='Paired')
```
